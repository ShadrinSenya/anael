```python
# -*- coding: utf-8 -*-
"""
MDK.05.02 Razrabotka koda informacionnyh sistem
Tema: Sozdanie nejrosetevogo proekta (Klassifikaciya emocij FER)
"""

# 1. IMPORT BIBLIOTEK
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import cv2
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import fetch_openml

import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models, callbacks
from tensorflow.keras.utils import to_categorical

import os
import urllib.request
import io
import warnings
warnings.filterwarnings('ignore')

# Ustanavlivaem seed dlya vosproizvodimosti
tf.random.set_seed(42)
np.random.seed(42)

print("Библиотеки загружены. Версия TF:", tf.__version__)

# 2. ZAGRUZKA DANNYH
def zagruzit_fer2013_iz_openml():
    print("\n[1] Загрузка датасета FER2013 через OpenML...")
    try:
        nabor_fer2013 = fetch_openml(data_id=40981, as_frame=True, parser='pandas')
        vse_dannye = nabor_fer2013.data
        vse_metki = nabor_fer2013.target

        spisok_izobrazhenij = []
        for stroka_pikselej in vse_dannye['pixels']:
            massiv_pikselej = np.array(stroka_pikselej.split(), dtype=np.float32)
            kartinka = massiv_pikselej.reshape(48, 48, 1)
            spisok_izobrazhenij.append(kartinka)

        X_vhod = np.array(spisok_izobrazhenij, dtype='float32') / 255.0
        y_metki = np.array(vse_metki, dtype='int')
        y_kategorii = to_categorical(y_metki, num_classes=7)

        print(f"Данные загружены. Размер X: {X_vhod.shape}")
        return X_vhod, y_kategorii, y_metki
    except Exception as oshibka:
        print(f"Ошибка OpenML: {oshibka}. Пробуем GitHub...")
        return zagruzit_fer2013_iz_github()

def zagruzit_fer2013_iz_github():
    print("\nЗагрузка данных с GitHub...")
    adres_url = "https://raw.githubusercontent.com/tech2etc/Build-Face-Emotion-Detection/main/fer2013.csv"
    try:
        otvet_servera = urllib.request.urlopen(adres_url)
        tablica_df = pd.read_csv(io.StringIO(otvet_servera.read().decode('utf-8')))
        stolbec_pikselej = tablica_df['pixels'].values
        stolbec_emocij = tablica_df['emotion'].values

        spisok_kartinok = []
        for s_piksel in stolbec_pikselej:
            m_piksel = np.array(s_piksel.split(), dtype=np.float32)
            spisok_kartinok.append(m_piksel.reshape(48, 48, 1))

        X_vhod = np.array(spisok_kartinok, dtype='float32') / 255.0
        y_metki = np.array(stolbec_emocij, dtype='int')
        y_kategorii = to_categorical(y_metki, num_classes=7)
        return X_vhod, y_kategorii, y_metki
    except Exception as e_github:
        print(f"Ошибка GitHub: {e_github}. Создание синтетических данных...")
        kolvo_primerov = 1000
        X_synth = np.random.rand(kolvo_primerov, 48, 48, 1).astype(np.float32)
        y_synth = np.random.randint(0, 7, kolvo_primerov)
        return X_synth, to_categorical(y_synth, 7), y_synth

# Poluchenie dannyh
X, y_cat, y_orig = zagruzit_fer2013_iz_openml()
X_obuch, X_test, y_obuch, y_test = train_test_split(X, y_cat, test_size=0.2, random_state=42)

# 3. MODEL
def sozdat_model():
    arhitektura = models.Sequential([
        layers.Input(shape=(48, 48, 1)),
        layers.Conv2D(32, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Flatten(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.5),
        layers.Dense(7, activation='softmax')
    ])
    return arhitektura

nejroset = sozdat_model()
nejroset.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 4. OBUCHENIE
print("\nНачало процесса обучения...")
zhurnal_obucheniya = nejroset.fit(X_obuch, y_obuch, epochs=10, batch_size=64, validation_split=0.2, verbose=1)

# 5. VIZUALIZACIYA
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(zhurnal_obucheniya.history['accuracy'], label='Tochnost (obuch)')
plt.plot(zhurnal_obucheniya.history['val_accuracy'], label='Tochnost (val)')
plt.title('График точности (Accuracy)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(zhurnal_obucheniya.history['loss'], label='Poteri (obuch)')
plt.plot(zhurnal_obucheniya.history['val_loss'], label='Poteri (val)')
plt.title('График функции потерь (Loss)')
plt.legend()
plt.show()

# 6. FUNKCIYA PREDSKAZANIYA
def predskazat_emociyu(put_k_fajlu):
    try:
        kartinka_bgr = cv2.imread(put_k_fajlu, cv2.IMREAD_GRAYSCALE)
        if kartinka_bgr is None: return "Файл не найден", 0
        kartinka_res = cv2.resize(kartinka_bgr, (48, 48))
        kartinka_final = kartinka_res.reshape(1, 48, 48, 1) / 255.0
        veroyatnosti = nejroset.predict(kartinka_final)
        nazvaniya_emocij = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']
        return nazvaniya_emocij[np.argmax(veroyatnosti)], np.max(veroyatnosti) * 100
    except Exception as e_pred:
        return f"Ошибка: {e_pred}", 0

nejroset.save('fer_model.keras')
print("\nВЫВОД: Модель успешно обучена и сохранена как 'fer_model.keras'.")
```

```python
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 7. MATRICA OSHIBOK
def postroit_matricu_oshibok(model_set, X_val, y_val):
    nazvaniya_emocij = ['Angry', 'Disgust', 'Fear', 'Happy', 'Neutral', 'Sad', 'Surprise']

    # Poluchaem predskazaniya
    y_pred_veroyatnosti = model_set.predict(X_val)
    y_pred_klassy = np.argmax(y_pred_veroyatnosti, axis=1)
    y_istinnye_klassy = np.argmax(y_val, axis=1)

    # Vychislyaem matricu
    matrica = confusion_matrix(y_istinnye_klassy, y_pred_klassy)

    # Vizualizaciya
    plt.figure(figsize=(10, 8))
    sns.heatmap(matrica, annot=True, fmt='d', cmap='Blues',
                xticklabels=nazvaniya_emocij,
                yticklabels=nazvaniya_emocij)
    plt.title('Матрица ошибок (Confusion Matrix)')
    plt.ylabel('Истинные метки')
    plt.xlabel('Предсказанные метки')
    plt.show()

# Zapusk vizualizacii
postroit_matricu_oshibok(nejroset, X_test, y_test)
```

```python
# 8. PRIMER ISPOLZOVANIYA FUNKCII PREDSKAZANIYA

# Berem sluchajnoe izobrazhenie iz testa dlya demonstracii
index_primera = np.random.randint(0, len(X_test))
kartinka_primera = X_test[index_primera].squeeze()

# Sohranyaem kartinku vo vremennyj fajl (imitaciya zagruzki polzovatelem)
put_fajla = 'test_face.png'
plt.imsave(put_fajla, kartinka_primera, cmap='gray')

# Vyzyvaem funkciyu predskazaniya
emociya, uverennost = predskazat_emociyu(put_fajla)

# Vyvod rezul'tata
print(f"--- Rezultat klasifikacii ---")
print(f"Fajl: {put_fajla}")
print(f"Predskazannaya emociya: {emociya}")
print(f"Uverennost: {uverennost:.2f}%")

# Pokazyvaem samo izobrazhenie
plt.imshow(kartinka_primera, cmap='gray')
plt.title(f"Predskazanie: {emociya} ({uverennost:.1f}%)")
plt.axis('off')
plt.show()
```

### 9. Uluchshenie modeli (Uglublennaya arhitektura i Augmentaciya)

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 1. Sozdanie uluchshennoj arhitektury
def sozdat_uluchshennuyu_model():
    model = models.Sequential([
        layers.Input(shape=(48, 48, 1)),

        # Blok 1
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(64, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Blok 2
        layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.Conv2D(128, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        # Blok 3
        layers.Conv2D(256, (3, 3), padding='same', activation='relu'),
        layers.BatchNormalization(),
        layers.MaxPooling2D((2, 2)),
        layers.Dropout(0.25),

        layers.Flatten(),
        layers.Dense(512, activation='relu'),
        layers.BatchNormalization(),
        layers.Dropout(0.5),
        layers.Dense(7, activation='softmax')
    ])
    return model

uluchshennaya_model = sozdat_uluchshennuyu_model()
uluchshennaya_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# 2. Nastrojka augmentacii (iskusstvennoe uvelichenie vyborki)
generator_dannyh = ImageDataGenerator(
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

print("Улучшенная архитектура создана. Готова к обучению с аугментацией.")
```

```python
# 3. Obuchenie uluchshennoj modeli
print("\nNachalo obucheniya uluchshennoj modeli...")

# Ispol'zuem ranee sozdannye X_obuch i y_obuch
zhurnal_uluchsheniya = uluchshennaya_model.fit(
    generator_dannyh.flow(X_obuch, y_obuch, batch_size=32),
    epochs=15, # Uvelichim kolichestvo epoh
    validation_data=(X_test, y_test),
    verbose=1
)

print("\nObuchenie zaversheno.")
```

### 10. Vizualizaciya rezul'tatov uluchshennoj modeli

```python
# Vizualizaciya processa obucheniya uluchshennoj modeli
plt.figure(figsize=(12, 4))

# Grafik tochnosti
plt.subplot(1, 2, 1)
plt.plot(zhurnal_uluchsheniya.history['accuracy'], label='Tochnost (obuch)', marker='o')
plt.plot(zhurnal_uluchsheniya.history['val_accuracy'], label='Tochnost (val)', marker='x')
plt.title('Uluchshennaya model: Accuracy')
plt.xlabel('Epohi')
plt.ylabel('Tochnost')
plt.legend()
plt.grid(True)

# Grafik poter
plt.subplot(1, 2, 2)
plt.plot(zhurnal_uluchsheniya.history['loss'], label='Poteri (obuch)', marker='o')
plt.plot(zhurnal_uluchsheniya.history['val_loss'], label='Poteri (val)', marker='x')
plt.title('Uluchshennaya model: Loss')
plt.xlabel('Epohi')
plt.ylabel('Poteri')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
